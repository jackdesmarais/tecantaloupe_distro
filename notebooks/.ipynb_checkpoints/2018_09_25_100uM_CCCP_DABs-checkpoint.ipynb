```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn
import itertools

from growth.plate_spec import PlateSpec
from growth.plate_time_course_parser import SavageLabM1000Excel
from matplotlib import pyplot as plt
```

```python
# PlateSpec defined in CSV for easy editing
ps = PlateSpec.FromFile('growth/plate_specs/2018_09_24_CCCP_100uM_pFE-DABs.csv')
name_mapping = ps.well_to_name_mapping()

# PlateSpec contains a DataFrame with the mapping. 
ps.df
```

```python
parser = SavageLabM1000Excel()
timecourse = parser.ParseFromFilename('growth/data/2018_09_25_100uM_CCCP_DABs.xlsx')

blanked = timecourse.blank()
# You can blank each well against it's own early timepoints
# by not setting blank_wells

# Smooth with default params.
smoothed = blanked.smooth()
```

```python
# Smoothed & blanked data.
fig = plt.figure(figsize=(10,10))
OD_data = smoothed.data_for_label('abs600')
OD_data.plot(x='time_s', legend=False)
plt.show()
```

```python
means = smoothed.mean_by_name(ps)
sems = smoothed.sem_by_name(ps)

mean_OD = means.data_for_label('abs600')
sems_OD = sems.data_for_label('abs600')
```

```python
# groupings of samples for plotting
pos = [c for c in mean_OD.columns
       if c.startswith('CAF+HCA')]
bw = [c for c in mean_OD.columns
       if c.startswith('BW')]
neg = [c for c in mean_OD.columns
       if c.startswith('CAF+G')]
exp_DAB = [c for c in mean_OD.columns
       if c.startswith('CAF+DAB')]
exp_sbta = [c for c in mean_OD.columns
       if c.startswith('CAF+sb')]
# exp_metal = [c for c in mean_OD.columns
#        if (c.startswith('CAF+pFE-DAB2') and c.endswith('atc'))]
# exp_metal_un = [c for c in mean_OD.columns
#        if (c.startswith('CAF+pFE-DAB2') and not c.endswith('atc'))]
```

```python

lag_times = means.LagTime(density_label='abs600')

labels = sorted(pos) + sorted(exp_growth) + sorted(neg)
lags = [lag_times[l] for l in labels]

# Note - samples that don't cross the threshold get a 0 lag value.
figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, lags)
plt.ylabel('Lag Time (hrs)')
plt.xticks(xs + 0.4, labels, rotation=90)
plt.show()
```

```python
growth_rates = means.MaxGrowthRates(density_label='abs600')

labels = sorted(pos) + sorted(exp_growth) + sorted(neg)
grs = [growth_rates[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, grs)
plt.ylabel('Max Growth Rate (/hr)')
plt.xticks(xs + 0.4, labels, rotation=90)
plt.show()
```

```python
yields = means.GrowthYield(density_label='abs600')

labels = sorted(pos) + sorted(exp_growth) + sorted(neg)
ylds = [yields[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(ylds))
plt.bar(xs, ylds)
plt.ylabel('Growth Yield (OD600 Units)')
plt.xticks(xs + 0.4, labels, rotation=90)
plt.show()
```

```python
yld = smoothed.GrowthYield(density_label='abs600')
name2well = ps.name_to_well_mapping()

yld = pd.DataFrame(yld, index=['yield']).T
grouped_yld = yld.groupby(ps.well_to_name_mapping())
mean_yield = grouped_yld.mean()
std_yield = grouped_yld.std()

mean_yield.plot(kind='bar', yerr=std_yield, figsize=(20,10), legend=False)
```

```python
yields = means.GrowthYield(density_label='abs600')

to_plot = [
    ('bw+pFE-sfGFP', 'WT vector (+)'),
    ('CAF+pFE-sfGFP', 'vector (-)'),
    ('CAF+pFE-HCA2', 'HCAII (+)'),
    ('CAF+pFE-DAB2_wt', 'DAB2'),
    ('CAF+pFE-DUF', 'DUF'),
    ('CAF+pFE-NDH', 'NDH'),
    ('CAF+pFE-DAB2+dCys1', 'NDH+DUF C539'),
    ('CAF+pFE-DAB2+dHis1', 'NDH+DUF H524'),
    
]

labels = [t[1] for t in to_plot]
ylds = [mean_yield.loc[t[0]][0] for t in to_plot]
sem = [std_yield.loc[t[0]][0] for t in to_plot]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(ylds))
plt.bar(xs, ylds, yerr=sem,capsize=5)
plt.ylabel('Growth Yield (OD600 Units)', fontsize=20)
plt.xticks(xs, labels, rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('./qualsYieldsFig.svg',format='svg')
plt.show()
```
