```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn
import itertools

from growth.plate_spec import PlateSpec
from growth.plate_time_course_parser import SavageLabM1000Excel
from matplotlib import pyplot as plt
```

```python
# PlateSpec defined in CSV for easy editing
ps = PlateSpec.FromFile('growth/plate_specs/2020_03_05_FolA_library_QC.csv')
name_mapping = ps.well_to_name_mapping()

# PlateSpec contains a DataFrame with the mapping. 
ps.df
```

```python
parser = SavageLabM1000Excel()
timecourse = parser.ParseFromFilename('growth/data/2020_03_05_folA_library_cy1_cy3_QC_24hours.xlsx')

blanked = timecourse.blank()
# You can blank each well against it's own early timepoints
# by not setting blank_wells

# Smooth with default params.
smoothed = blanked.smooth()
```

```python
# Smoothed & blanked data.
fig = plt.figure(figsize=(10,10))
OD_data = smoothed.data_for_label('abs600')
OD_data.plot(x='time_s', legend=False)
plt.show()
```

```python
means = smoothed.mean_by_name(ps)
sems = smoothed.sem_by_name(ps)

mean_OD = means.data_for_label('abs600')
sems_OD = sems.data_for_label('abs600')
```

```python
# groupings of samples for plotting
cy3_supps = [c for c in mean_OD.columns
       if (c.startswith('Suppliments_3Cycle') or c.startswith('Suppliments_WT'))]
cy1_supps = [c for c in mean_OD.columns
       if (c.startswith('Suppliments_1Cycle') or c.startswith('Suppliments_WT'))]
cy3_thyA = [c for c in mean_OD.columns
       if (c.startswith('thyA_sup_3Cycle') or c.startswith('thyA_WT'))]
cy1_thyA = [c for c in mean_OD.columns
       if (c.startswith('thyA_sup_1Cycle') or c.startswith('thyA_WT'))]
```

```python
seaborn.set_style('white')
pos_to_plot = cy3_supps
neg_to_plot = cy3_thyA

plt.figure(figsize=(20,10))
time_h = mean_OD.time_s / (60.0*60.0)

colors = seaborn.color_palette()
for i, c in enumerate(pos_to_plot):
    label = c
    color = colors[i % len(colors)]
    plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
                     color=color, alpha=0.5)
    plt.plot(time_h, mean_OD[c], label=label, color=color)

for i, c in enumerate(neg_to_plot):
    label = c
    color = colors[i % len(colors)]
    plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
                     color=color, alpha=0.5)
    plt.plot(time_h, mean_OD[c], label=label, color=color,ls='--')
    
# for i, c in enumerate(exp_to_plot):
#     label = c
#     color = colors[i % len(colors)]
#     plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
#                      color=color, alpha=0.5)
#     plt.plot(time_h, mean_OD[c], label=label, color=color,ls='-.')
plt.legend(loc=(1,0), fontsize=18)
plt.xlim(0, np.max(time_h)+.1)
plt.ylim(-0.1, 1.3)
plt.xticks(np.arange(0, np.max(time_h)+.1, 5), fontsize=14)
plt.yticks(fontsize=14)
plt.title('Cy3 library', fontsize=24)
plt.xlabel('Time (Hours)', fontsize=20)
plt.ylabel('OD600', fontsize=20)

plt.show()
```

```python
# groupings of samples for plotting
cy3_full = [c for c in mean_OD.columns
       if c.startswith('Suppliments_3Cycle')]
cy1_full = [c for c in mean_OD.columns
       if c.startswith('Suppliments_1Cycle')]
cy3_thy = [c for c in mean_OD.columns
       if c.startswith('thyA_sup_3Cycle')]
cy1_thy = [c for c in mean_OD.columns
       if c.startswith('thyA_sup_1Cycle')]
wt = [c for c in mean_OD.columns
       if (c.startswith('Suppliments_WT') or c.startswith('thyA_WT'))]


lag_times = means.LagTime(density_label='abs600')
wt_labels = sorted(wt)
pos_labels = sorted(cy3_full)
neg_labels = sorted(cy3_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
lags = [lag_times[l] for l in labels]


neg_lags = np.array([lag_times[l] for l in neg_labels])
pos_lags = np.array([lag_times[l] for l in pos_labels])
f = len(neg_lags[~np.isnan(neg_lags)])/len(pos_lags[~np.isnan(pos_lags)])

# Note - samples that don't cross the threshold get a 0 lag value.
figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, lags, color=colors)
plt.ylabel('Lag Time (hrs)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy3 library\nFvalue: %s'%f, fontsize=24)
plt.show()

wt_labels = sorted(wt)
pos_labels = sorted(cy1_full)
neg_labels = sorted(cy1_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
lags = [lag_times[l] for l in labels]


neg_lags = np.array([lag_times[l] for l in neg_labels])
pos_lags = np.array([lag_times[l] for l in pos_labels])
f = len(neg_lags[~np.isnan(neg_lags)])/len(pos_lags[~np.isnan(pos_lags)])

# Note - samples that don't cross the threshold get a 0 lag value.
figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, lags, color=colors)
plt.ylabel('Lag Time (hrs)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy1 library\nFvalue: %s'%f, fontsize=24)

plt.show()
```

```python
growth_rates = means.MaxGrowthRates(density_label='abs600')

wt_labels = sorted(wt)
pos_labels = sorted(cy3_full)
neg_labels = sorted(cy3_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
grs = [growth_rates[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, grs, color=colors)
plt.ylabel('Max Growth Rate (/hr)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy3 library', fontsize=24)

plt.show()

wt_labels = sorted(wt)
pos_labels = sorted(cy1_full)
neg_labels = sorted(cy1_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
grs = [growth_rates[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(lags))
plt.bar(xs, grs, color=colors)
plt.ylabel('Max Growth Rate (/hr)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy1 library', fontsize=24)
plt.show()
```

```python
yields = means.GrowthYield(density_label='abs600')

wt_labels = sorted(wt)
pos_labels = sorted(cy3_full)
neg_labels = sorted(cy3_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
ylds = [yields[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(ylds))
plt.bar(xs, ylds, color=colors)
plt.ylabel('Growth Yield (OD600 Units)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy3 library', fontsize=24)
plt.show()

wt_labels = sorted(wt)
pos_labels = sorted(cy1_full)
neg_labels = sorted(cy1_thy)
labels = wt_labels + pos_labels + neg_labels
colors= ['r']* len(wt_labels) + ['b']*len(pos_labels)+['g']*len(neg_labels)
ylds = [yields[l] for l in labels]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(ylds))
plt.bar(xs, ylds, color=colors)
plt.ylabel('Growth Yield (OD600 Units)', fontsize=20)
plt.xticks(xs + 0.4, labels, rotation=90)
plt.title('cy1 library', fontsize=24)
plt.show()
```

```python
mutations = pd.read_csv('./growth/data/2020_03_05_folA_library_cy1_cy3_QC_mutations.csv')
```
