```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn
import itertools

from growth.plate_spec import PlateSpec
from growth.plate_time_course_parser import SavageLabM1000Excel
from matplotlib import pyplot as plt
```

```python
# PlateSpec defined in CSV for easy editing
ps = PlateSpec.FromFile('growth/plate_specs/2018_03_23_operon1CAFreeKnockInGrowth_metalandproton_nice.csv')
name_mapping = ps.well_to_name_mapping()

# PlateSpec contains a DataFrame with the mapping. 
ps.df
```

```python
parser = SavageLabM1000Excel()
timecourse = parser.ParseFromFilename('growth/data/2018_03_24_CAFree_DAB2_protonWire.xlsx')

blanked = timecourse.blank()
# You can blank each well against it's own early timepoints
# by not setting blank_wells

# Smooth with default params.
smoothed = blanked.smooth()
```

```python
# Smoothed & blanked data.
fig = plt.figure(figsize=(10,10))
OD_data = smoothed.data_for_label('abs600')
OD_data.plot(x='time_s', legend=False)
plt.show()
```

```python
means = smoothed.mean_by_name(ps)
sems = smoothed.sem_by_name(ps)

mean_OD = means.data_for_label('abs600')
sems_OD = sems.data_for_label('abs600')
```

```python
# groupings of samples for plotting
pos = [c for c in mean_OD.columns
       if c.startswith('bw') or c.startswith('CAF+pFE-DAB2_wt')]
neg = [c for c in mean_OD.columns
       if c.startswith('CAF+pFE-sf')]
exp_zinc = [c for c in mean_OD.columns
       if c.startswith('CAF+pFE-DC') or c.startswith('CAF+pFE-DD')]
exp_proton_ind = [c for c in mean_OD.columns
       if (c.startswith('CAF+pFE-N') and c.endswith('atc'))]
exp_proton_u = [c for c in mean_OD.columns
       if (c.startswith('CAF+pFE-N') and not c.endswith('atc'))]
```

```python

# to_plot = [
#     ('bw+pFE-sfGFP+atc', 'WT vector (+)'),
#     ('CAF+pFE-sfGFP+atc', 'vector (-)'),
#     ('CAF+pFE-HCA2+atc', 'HCAII (+)'),
#     ('CAF+pFE-DAB2_wt+atc', 'DAB2'),
#     ('CAF+pFE-DAB2+dAsp1+atc', 'DAB2 D538A'),
#     ('CAF+pFE-DAB2+dCys1+atc', 'DAB2 C539A'),
#     ('CAF+pFE-DAB2+dCys2+atc', 'DAB2 C542A'),
#     ('CAF+pFE-DAB2+dCys_1&2+atc', 'DAB2 2C2A'),
#     ('CAF+pFE-DAB2+dHis1+atc', 'DAB2 H524A'), 
#     ('CAF+pFE-DAB2+dHis2+atc', 'DAB2 H533A')]
```

```python
seaborn.set_style('white')
pos_to_plot = pos
exp_to_plot = exp_zinc
neg_to_plot = neg

plt.figure(figsize=(20,10))
time_h = mean_OD.time_s / (60.0*60.0)

colors = seaborn.color_palette()
for i, c in enumerate(pos_to_plot):
    label = c
    color = colors[i % len(colors)]
    plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
                     color=color, alpha=0.5)
    plt.plot(time_h, mean_OD[c], label=label, color=color)

for i, c in enumerate(neg_to_plot):
    label = c
    color = colors[i % len(colors)]
    plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
                     color=color, alpha=0.5)
    plt.plot(time_h, mean_OD[c], label=label, color=color,ls='--')
    
for i, c in enumerate(exp_to_plot):
    label = c
    color = colors[i % len(colors)]
    plt.fill_between(time_h, mean_OD[c] - sems_OD[c], mean_OD[c] + sems_OD[c],
                     color=color, alpha=0.5)
    plt.plot(time_h, mean_OD[c], label=label, color=color,ls='-.')
plt.legend(loc=2, fontsize=18)
plt.xlim(0, np.max(time_h)+.1)
plt.ylim(-0.1, 1.3)
plt.xticks(np.arange(0, np.max(time_h)+.1, 5), fontsize=14)
plt.yticks(fontsize=14)
plt.title('Growth  metal binders', fontsize=20)
plt.xlabel('Time (Hours)', fontsize=20)
plt.ylabel('OD600', fontsize=20)

plt.show()
```

```python
yld = smoothed.GrowthYield(density_label='abs600')
yld = pd.DataFrame(yld, index=['yield']).T
grouped_yld = yld.groupby(name_mapping)
mean_yield = grouped_yld.mean()
std_yield = grouped_yld.std()

to_plot =  [ ('bw+pFE-sfGFP', 'WT vector (+)'),
                ('CAF+pFE-DAB2_wt', 'CAF DAB2 (+)'),
                ('CAF+pFE-DC351', 'DC351'),
                ('CAF+pFE-DC351+atc', 'DC351 induced'),
                ('CAF+pFE-DD353', 'DD353'),
                ('CAF+pFE-DD353+atc', 'DD353 induced'),
                ('CAF+pFE-NH339+atc', 'NH339 induced'),
                ('CAF+pFE-NH339', 'NH339'),
                ('CAF+pFE-NK347+atc', 'NK347 induced'),
                ('CAF+pFE-NK347', 'NK347'),
               ('CAF+pFE-sfGFP', 'CAF vector (-)'),
              ]

labels = [t[1] for t in to_plot]
ylds = [mean_yield.loc[t[0]].values[0] for t in to_plot]
stds = [std_yield.loc[t[0]].values[0] for t in to_plot]

figure = plt.figure(figsize=(10,10))
xs = np.arange(len(ylds))
plt.bar(xs, ylds,yerr=stds)
plt.ylabel('Growth Yield (OD600 Units)', fontsize=20)
plt.xticks(xs, labels, rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
